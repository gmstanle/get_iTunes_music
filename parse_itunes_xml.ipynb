```python
import plistlib
import pandas as pd
```

```python
libpath = '/Users/geoffstanley/Music/iTunes/iTunes Music Library.xml'
```

```python
plist=plistlib.readPlist(libpath)
```

```python
track
```

```python
tracklist=[]
artistlist=[]
albumlist=[]
trackidlist=[]
for i, x in enumerate(plist['Tracks']):
    track=plist['Tracks'].get(x)
    tracklist.append(track.get('Name'))
    artistlist.append(track.get('Artist'))
    albumlist.append(track.get('Album'))
    trackidlist.append(track.get('Track ID'))
```

```python
track_df = pd.DataFrame(list(zip(tracklist,artistlist,albumlist,trackidlist)), 
                        columns=['Track','Artist','Album','TrackID'])
```
